#### 1. Пользователь в бесконечном цикле вводит числа через консоль до тех пор, пока не введёт что-то, отличное от числа (после этого цикл завершается). В конце программа должна вывести сумму всех введённых чисел.

```python
sum_total = 0
while True:
    number = input('Enter any number : ')
    try:
        sum_total += float(number)
        print(f'Number entered = {number}')
    except ValueError:
        break
print(f'Sum of the entered numbers = {sum_total}')
```

#### 2. Пользователь вводит строку через консоль (состоящую только из букв в разном регистре). Программа должна создать новую строку, содержащую только те буквы из исходной строки, которые были в верхнем регистре.

```python
upper_letter = []
text = input('Enter your letters : ')
print(f'Entered your letters :{text}')
for letter in text:
    if letter.isupper():
        upper_letter.append(letter)
print(f'Entered upper letters :{''.join(upper_letter)}')
```

#### 3. Классическая игра FizzBuzz. Программа в цикле выводит в консоль числа от 1 до 100. Если число кратно 3, вместо него выводится Fizz. Если кратно 5 — Buzz. Если кратно и 3, и 5 — FizzBuzz.

```python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print(f'Number = {i}, FizzBuzz')
    elif i % 3 == 0:
        print(f'Number = {i}, Fizz')
    elif i % 5 == 0:
        print(f'Number = {i}, Buzz')
    else:
        print(f'Number = {i}')
```

## Задание 1: Зеркальное отражение цифр

Пользователь вводит пятизначное число. Программа должна зеркально отразить три центральные цифры. Первая и последняя остаются на местах.

<br>

**Примеры:**
* `23456 -> 25436`
* `30789 -> 38709`

```python
num = input('Enter a 5-digit number: ')
if len(num) == 5 and num.isdigit():
    transformed = num[0] + num[3] + num[2] + num[1] + num[4]
    print(f'Your number is {num}')
    print(f'Transformed number is {transformed}')
else:
    print(f'Error! You should have entered 5 digits!')
```

## Задание 2: Подсчет выходных до отпуска

Пользователь вводит сколько дней осталось до ближайшего отпуска. Программа должна вывести количество выходных дней до отпуска, если учесть, что выходные это суббота и воскресенье, сегодня понедельник и праздники мы не учитываем.  

<br>

**Примеры:**
* `4  -> 0`
* `6  -> 1`
* `14 -> 4`

```python
try:
    days_before_vacation = int(input('Enter the number of days until your vacation: '))
    print(f'Calendar days before your vacation: {days_before_vacation}')
    if days_before_vacation < 0:
        print("Error! The number of days cannot be negative.")
    else:
        weekend_count = 0
        for i in range(days_before_vacation):
            weekday = i % 7
            if weekday >= 5:
                weekend_count += 1
        print(f'Number of days off before your vacation: {weekend_count}')
except ValueError:
    print('Error! Enter a positive integer.')
```

## Задание 3: Плитка шоколада

Пользователь вводит длину и ширину плитки шоколада, а также размер куска, который хочет отломить. Программа должна вычислить - можно ли совершить подобный разлом или нет, если учесть, что ломать можно только по прямой.

<br>

**Примеры:**
* `3, 4, 6  -> True`
* `5, 7, 8  -> False`
* `4, 5, 12 -> True`

```python
try:
    length = int(input('Enter the length of the chocolate bar (number of pieces): '))
    width = int(input('Enter the width of the chocolate bar (number of pieces): '))
    size = int(input('Enter the size of the piece you want to break off (number of pieces): '))
except ValueError:
    print('Error! Enter a positive integer.')
    exit()
print(f'Length of the chocolate bar: {length}')
print(f'Width of the chocolate bar: {width}')
print(f'Size of the piece of the chocolate bar: {size}')

if 0 < size < length * width and (size % length == 0 and 
                               size // length < width or 
                               size % width == 0 and 
                               size // width < length):
    print('It is POSSIBLE to break off such a piece with one straight break!')
else:
    print('It is IMPOSSIBLE to break off such a piece with one straight break...')
```

## Задание 4: Римские числа

Пользователь вводит целое положительное число, программа должна вернуть строку в виде римского числа.

<br>

**Примеры:**
* `3   -> III`
* `15  -> XV`
* `234 -> CCXXXIV`

```python
roman_symbols = [
    (1000, 'M'), (900, 'CM'), (500, 'D'), (400, 'CD'),
    (100, 'C'), (90, 'XC'), (50, 'L'), (40, 'XL'),
    (10, 'X'), (9, 'IX'), (5, 'V'), (4, 'IV'), (1, 'I')
]

try:
    num = int(input('Enter a positive integer (1-3999): '))
    if not 1 <= num <= 3999:
        print('Error! The number must be between 1 and 3999.')
    else:
        roman_result = []
        temp = num
        for value, symbol in roman_symbols:
            while temp >= value:
                roman_result.append(symbol)
                temp -= value
        roman_digit = ''.join(roman_result)
        print(f'Entered number :{num}')
        print(f'Roman numeral representation of the entered number: {roman_digit}')
except ValueError:
    print('Error! Enter positive integer.')
```

## Задание 5: Проверка на вещественное число

Пользователь вводит данные, проверить - являются ли они положительным вещественным числом. Не использовать встроенные функции для проверки, только методы данных и конструкцию IF. (Дополнительное задание, по желанию - проверка на отрицательные вещественные числа)

<br>

**Примеры:**
* `5.6  -> True`
* `.78  -> True`
* `.67. -> False`
* `5    -> True`

```python
numder = input('Enter any number : ')

try:
    num = float(numder)    
    if num == num and num != float('inf') and num != float('-inf'):
        print(f'{numder} - is a real number.')
    else:
        print(f'{numder} - is not a real number.')
except ValueError:
    print(f'{numder} - is not a real number.')
```
