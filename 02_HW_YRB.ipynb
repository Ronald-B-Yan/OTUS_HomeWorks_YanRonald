#### 1. Написать программу, которая получает на вход строку и возвращает словарь, где: ключи — символы из этой строки; значения — количество раз, сколько каждый символ встречается.

#### 3. Написать программу, которая удаляет из списка все дубликаты, сохранив исходный порядок элементов

````python
list_items = input('Enter list items separated by spaces:')
items = list_items.split()
seen = set()
unique_items = []
for item in items:
    if item not in seen:
        seen.add(item)
        unique_items.append(item)
print(f'Your entered list: {list_items}')
print(f'List without duplicates: {unique_items}')
````

#### 4. Даны три (или больше) списка с объектами. Программа должна создать новый список, содержащий все уникальные элементы — каждый объект встречается только один раз.

````python
# Requesting the number of lists (at least 3)
while True:
    try:
        number_lists = int(input('Enter the number of lists (at least 3): '))
        if number_lists >= 3:
            break
        print('Error: You must enter a number >= 3.')
    except ValueError:
        print('Error: Enter an integer.')

# Creating lists with elements
all_items = []
for i in range(1, number_lists + 1): 
    raw = input(f'Enter the elements of the {i}-th list separated by spaces: ')
    print(f'The elements entered on the {i}-th list: {raw}')
    parts = raw.strip().split()
    converted = []
    for j in parts:
            converted.append(j)
    all_items.extend(converted)

# Choosing unique elements
unique_items = []
seen = set()
for item in all_items:
    if item not in seen:
        unique_items.append(item)
        seen.add(item)

print(f'Your entered elements: {all_items}')
print(f'Resulting list of unique elements: {unique_items}')
````

#### 4. Дана строка текста (или введённая через консоль). Программа должна вернуть словарь с четырьмя ключами: "гласные", "согласные", "цифры", "пунктуация". Значения — количество символов каждого типа в строке.

````python
text = input('Enter your text: ')

# Creating a vowels list
vowels_ru = 'аеёиоуыэюя'
vowels_en = 'aeiouy'
vowels = set(vowels_ru + vowels_en)

# Creating a list of punctuation
punctuation = r"""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""

# Creating an empty dictionary
counts = {
    'vowels': 0,
    'consonants': 0,
    'numbers': 0,
    'punctuation': 0
}

# Text analysis
for char in text:
    if char.isdigit():
        counts["numbers"] += 1
    elif char.isalpha():
        if char.lower() in vowels:
            counts["vowels"] += 1
        else:
            counts["consonants"] += 1
    elif char in punctuation:
        counts["punctuation"] += 1

print(f'Your entered text: {text}')
print(f'Your text consists of: {counts}')
````

## Задание 1: Получение однозначного числа

Пользователь вводит целое число, программа складывает все цифры числа, с полученным числом — то же самое, и так до тех пор, пока не получится однозначное число.

<br>

**Примеры:**
* `545   -> 5`
* `12345 -> 6`

````python
num = int(input('Enter any integer digit: '))
abs_num = abs(num)

while abs_num >= 10:
    single_digit = 0
    temp = abs_num
    while temp > 0:
        single_digit += temp % 10
        temp //= 10
    abs_num = single_digit

print(f'Your entered digit: {num}')
print(f'Single-digit number of the entered number: {single_digit}')
````

## Задание 2: Кинотеатр

Дан список списков, каждый вложенный список состоит из 1 и 0, количество вложенных списков — количество рядов. Пользователь вводит, сколько билетов ему требуется. Программа должна найти ряд, где можно приобрести нужное количество билетов (места должны быть рядом). Если таких рядов несколько, то ближайший к экрану (ближайшим считается нулевой ряд). Если таких мест нет, то вывести False.

<br>

**Примеры:**
* `[[0,1,1,0], [1,0,0,0], [0,1,0,0]], 2 -> 1`
* `[[0,1,1,0], [1,0,1,0], [1,1,0,1]], 2 -> False`

````python
# Creation of a cinema hall
# 0 - free, 1 - occupied
cinema = [
    [0, 1, 0, 0, 1],           # Row 0 (closest to the screen)
    [1, 0, 0, 1, 0, 0, 1],     # Row 1
    [0, 1, 1, 0, 0, 1, 1],     # Row 2
    [0, 0, 0, 0, 0, 1, 1, 1],  # Row 3
    [0, 0, 1, 1, 1, 0, 0, 1]   # Row 4
]

tickets = int(input('How many tickets do you need?'))

result = False
if tickets > 0:
    for i in range(len(cinema)):
        row = cinema[i]
        consecutive_free = 0
        for seat in row:
            if seat == 0:
                consecutive_free += 1
                if consecutive_free >= tickets:
                    result = i
                    break
            else:
                consecutive_free = 0
        if result is not False:
            break

print(f'You need {tickets} tickets')
print(f'{tickets} tickets can be purchased in row {result}')

# print(result)
````

## Задание 3: Алгоритм RLE

Необходимо написать упрощенную версию алгоритма RLE. 
Алгоритм RLE объединяет подряд идущие символы в коэффициент и символ.

<br>

**Примеры:**
* `asssdddsssddd -> 1a3s3d3s3d`
* `aaabbbbccccc  -> 3a4b5c`
* `abcba         -> 1a1b1c1b1a`

````python
string = input('Enter the string for RLE encoding:')
if not string:
    result = ""
else:
    encoded = []
    count = 1
    prev_char = string[0]
    for char in string[1:]:
        if char == prev_char:
            count += 1
        else:
            encoded.append(str(count) + prev_char)
            prev_char = char
            count = 1
    encoded.append(str(count) + prev_char)
    result = ''.join(encoded)

print(f'You string for encoding: {string}')
print(f'RLE encoding result: {result}')
````

## Задание 4: Шифр Цезаря

Реализуйте программу для шифрования текста с помощью шифра Цезаря.  
Шифр Цезаря — это метод шифрования, при котором каждая буква в тексте заменяется буквой, стоящей на фиксированное число позиций дальше в алфавите.

### Требования
1. Программа должна принимать на вход:
   - Строку текста для шифрования
   - Ключ шифра (число от 0 до 25)

2. Правила шифрования:
   - Шифруются только латинские буквы (a-z, A-Z)
   - Сохраняется регистр букв (заглавные остаются заглавными)
   - Пробелы, цифры и знаки препинания остаются без изменений
   - Сдвиг циклический (после 'z' идет 'a', после 'Z' идет 'A')

### Примеры

* `"Hello World", 3      -> "Khoor Zruog"`
* `"Python is great!", 5 -> "Udymts nx lwjfy!"`
* `"XYZ abc", 3          -> "ABC def"`
* `"Test 123", 1         -> "Uftu 123"`

### Алгоритм
* Для каждого символа в строке:
   - Если это латинская буква, найти её позицию в алфавите
   - Добавить к позиции значение ключа
   - Если результат больше 25, использовать остаток от деления на 26
   - Преобразовать обратно в букву, сохранив регистр
   - Если это не буква, оставить символ без изменений

````python
text = input('Enter the text to encrypt: ')

while True:
    try:
        key = int(input('Enter the encryption key (0-25): '))
        if key >= 0 and key <= 25:
            break
        print('Error: You must enter a number between 0 and 25.')
    except ValueError:
        print('Error: Enter an integer.')

result = []
for char in text:
    if 'a' <= char <= 'z':
        # Processing of lowercase letters
        shifted = ord('a') + (ord(char) - ord('a') + key) % 26
        result.append(chr(shifted))
    elif 'A' <= char <= 'Z':
        # Processing of capital letters
        shifted = ord('A') + (ord(char) - ord('A') + key) % 26
        result.append(chr(shifted))
    else:
        # Processing of non-letter characters
        result.append(char)
    encripted_text = ''.join(result)

print(f'Your text to encrypt: {text}')
print(f'Your encryption key: {key}')
print(f'Encrypted text: {encripted_text}')
````

## Задание 5: Табель успеваемости

Пользователь в бесконечном цикле (пока не будет введена пустая строка) вводит строки вида: 'название предмета' 'фамилия ученика' 'оценка'. После окончания ввода программа выводит в консоль название предмета, далее список учеников и все их оценки в виде таблицы.

<br>

**Примеры:**

<br>

*Ввод:*
```python
Математика Иванов  5
Математика Иванов  4
Литература Иванов  3
Математика Петров  5
Литература Сидоров 3
Литература Петров  5
Литература Иванов  4
Математика Сидоров 3
Математика Петров  5
```

<br>

*Результат:*
```python
Математика # вводим название предмета

# выводим список учеников и их оценки
Иванов  5 4
Петров  5 5
Сидоров 3
```

```python
Литература # вводим название предмета

# выводим список учеников и их оценки
Иванов  3 4
Сидоров 3
Петров  5
```

````python
data = {}
while True:
    line = input("Введите данные в формате 'Предмет Фамилия Оценка'. Для завершения введите пустую строку").strip()
    if not line:
        break

    parts = line.split()
    if len(parts) != 3:
        print('Ошибка: строка должна содержать ровно три части (Предмет Фамилия Оценка). Попробуйте снова.')
        continue

    subject, student, grade = parts
    data.setdefault(subject, {})
    data[subject].setdefault(student, [])
    data[subject][student].append(grade)

# Output of results
for subject, students in data.items():
    print(f'\nПредмет: {subject}')
    # Sorting students alphabetically
    for student in sorted(students.keys()):
        grades = students[student]
        print(f'{student} {' '.join(grades)}')
````
