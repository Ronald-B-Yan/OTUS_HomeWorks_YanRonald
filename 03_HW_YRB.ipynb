#### 1. Написать функцию, которая принимает неограниченное количество чисел в виде позиционных аргументов и ключевой аргумент — операцию над этими числами (сложение или умножение). Функция должна возвращать результат выполнения указанной операции.

```python
def calculate(*args, operation):

    if not args:
        raise ValueError('Missing input numbers')

    if operation == 'sum':
        return sum(args)

    if operation == 'multiply':
        result = 1
        for x in args:
            result *= x
        return result

    raise ValueError(f'Illegal operation: {operation}')

sum_total = calculate(1, 2, 3, 4, 5.02, 3e2, operation='sum')
print(f'Summation result: {sum_total}')

multiply_total = calculate(2, 3, 4, -5.5, 1e2, operation='multiply')
print(f'Multiplication result: {multiply_total}')
```

#### 3.Написать функцию, которая создаёт абсолютный путь к файлу. Позиционные аргументы:

название диска,

неограниченное количество папок,

имя файла (без расширения).


Ключевые аргументы:
ext — расширение файла,

sep — разделитель (по умолчанию '/').


Пример:
full_path('c:', 'work', 'python', 'function', 'main', ext='py') ➜ 'c:/work/python/function/main.py'

```python
def full_path(drive, *paths, ext='', sep='\\'):

    if not paths:
        raise ValueError('You must specify at least the file name')

    *folders, filename = paths
    path_parts = [drive] + folders
    dir_path = sep.join(path_parts)

    if ext:
        filename = f'{filename}.{ext}'

    return f'{dir_path}{sep}{filename}'

result = full_path('d:', 'my_work', 'python', 'function', 'main', ext='ipynb')
print(f'Full path to your file: {result}')
```

#### 4. Написать функцию, которая принимает список, состоящий из объектов разных типов, и возвращает словарь, где:
ключи — типы данных объектов;
значения — списки объектов соответствующего типа.

```python
def group_by_type(items):

    result = {}
    for object in items:
        object_type = type(object)
        if object_type not in result:
            result[object_type] = []
        result[object_type].append(object)
    return result
    
data = [42, 'Great', 3.14, [1, 2], 'Python', 'Hello World', 2.48e5, 100, {'a': 1, 'b': 3}, {'c': 8}]
dictionary = group_by_type(data)

for object_type, values in dictionary.items():
    print(f'{object_type.__name__}: {values}')
    
```

## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`

```python
def convert_case(row: str, output_format: str = 'auto'):

    input_format = 'snake' if '_' in row else 'pascal'
    target_format = 'pascal' if input_format == 'snake' else 'snake'
    print(f'There will be a transformation: {input_format} -> {target_format}')

    if target_format == 'pascal':
        words = row.split('_')
        return ''.join(word.capitalize() for word in words)

    result = []
    for i, char in enumerate(row):
        if char.isupper() and i !=0:
            prev_char = row[i-1]
            next_char = row[i+1] if i+1 < len(row) else None
            if prev_char.islower() or (prev_char.isupper() and next_char.islower()):
                result.append('_')
        result.append(char.lower())
    return ''.join(result)

result = convert_case('otus_course')
print(f'The result of the transformation: {result}\n')

result = convert_case('PythonIsTheBest')
print(f'The result of the transformation: {result}\n')

result = convert_case('OtusCourse')
print(f'The result of the transformation: {result}\n')

result = convert_case('python_is_the_best')
print(f'The result of the transformation: {result}')
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`

```python
def is_valid_date(date_str: str):

    try:
        parts = date_str.strip().split('.')
        if len(parts) != 3:
            return False
        
        day, month, year = map(int, parts)
        days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        
        if month < 1 or month > 12:
            return False

        if month == 2:
            is_leap = (year % 4 == 0)
            if is_leap:
                days_in_month[1] = 29
        
        if day < 1 or day > days_in_month[month - 1]:
            return False
        
        return True

    except (ValueError, TypeError):
        return False
        
your_date = input('Enter any date in the format DD.MM.YYYY: ')
print(f'You entered the date {your_date}. It\'s {is_valid_date(your_date)}.')
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
def is_prime(n: int):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
    
your_number = int(input('Enter any integer number: '))
print(f'You entered the number {your_number} to check if it\'s prime. It\'s {is_prime(your_number)}.\n')

your_number = int(input('Enter any integer number: '))
print(f'You entered the number {your_number} to check if it\'s prime. It\'s {is_prime(your_number)}.\n')

your_number = int(input('Enter any integer number: '))
print(f'You entered the number {your_number} to check if it\'s prime. It\'s {is_prime(your_number)}.')
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
def users_list():
    users = {}

    while True:
        first_name = input('Enter user\'s first name: ')
        if first_name == '':
            break
        if not first_name.isalpha():
            print('Error! The name must contain only letters.')
            continue
        first_name = first_name.capitalize()

        last_name = input('Enter user\'s last name: ')
        if not last_name.isalpha():
            print('Error! The last name must contain only letters.')
            continue
        last_name = last_name.capitalize()

        age = input('Enter user\'s age: ')
        if not age.isdigit() or not (18 <= int(age) <= 60):
            print('Error! Age must be between 18 and 60.')
            continue

        user_id = input('Enter user\'s ID: ')
        if not user_id.isdigit():
            print('Error! User\'s ID must be number.')
            continue
        user_id = user_id.zfill(8)

        if user_id in users:
            print('Error! User\'s ID must be unique.')
            continue

        users[user_id] = (first_name, last_name, age)
        print(f'User {last_name} {first_name} with ID {user_id} successfully added \n')

    print('User\'s list:')
    for user_id, (first_name, last_name, age) in users.items():
        print(f'User ID: {user_id} | First Name: {first_name} | Last Name: {last_name} | Age: {age}')

users_list()
```
